```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
validation = pd.read_csv('runs/Val_loss_GPU.csv',  encoding='ISO-8859-1',sep=',', error_bad_lines=False, header=None, skiprows=1)
val = []
training = pd.read_csv('runs/Train_loss_GPU.csv',  encoding='ISO-8859-1',sep=',', error_bad_lines=False, header=None, skiprows=1)
train = []
```

```python
for index, row in validation.iterrows():
    val.append(row[2])
```

```python
for index, row in training.iterrows():
    train.append(row[2])
```

```python
f = plt.figure()
plt.plot(train)
plt.plot(val)
plt.title('Loss')
plt.xlabel('Epochs')
plt.legend(['Train','Val'])

plt.grid()
plt.show()
```

```python
f.savefig("Loss_Plot.pdf")
```
